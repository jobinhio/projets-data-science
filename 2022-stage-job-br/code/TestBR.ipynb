```python
import codeEF as EF
import codeBR as BR 
import codemesh as MESH
import matplotlib.pyplot as plt
```

### Resolution BR avec $N_{0} = 2$ 

```python
# Maillage omega
Nel = 100
mh = MESH.mesh(Nel)

# classe fem
femP1 = EF.fem(mh)

#classe BR
N_mu = 100
br = BR.Br(femP1,N_mu)

#genere liste_mu et liste_mu_test
br.genere_liste_mu()
br.give_liste_mu_test()


# on applique l'algo glouton
N_0 =  2
br.Greedy(N_0)


# resolution du pb avec Ubr pour un mu fixé
mu= 1
Ubr = br.solveUbr(mu, plot=True)
```

L'objectif principal de mon stage était d'optimiser les coûts de production en déterminant la meilleure composition du lit de fusion. Cette tâche impliquait de trouver la proportion optimale de matières premières à utiliser lors du processus de fusion. Ces matières premières proviennent de diverses sources, incluant des matériaux métalliques et de construction, ainsi que des retours, c'est-à-dire des résidus des cycles de production précédents. Parmi ces matériaux, on compte par exemple des sabots de freins de la SNCF ou encore des rails de chemin de fer de 40 cm. Chaque matière première possède une composition chimique spécifique, et la qualité de la fonte produite dépend de cette composition, qui doit respecter des valeurs précises pour garantir le résultat souhaité. Pour atteindre cet objectif, j'ai implémenté deux approches utilisant tous deux l'algorithme du simplexe. Après avoir formulé le problème comme un programme linéaire avec des contraintes linéaires, j'ai développé deux approches distinctes : la première se concentre sur la réduction des coûts, tandis que la deuxième met l'accent sur l'amélioration de la qualité de la fonte, tout en minimisant l'impact sur les coûts.
Je vais maintenant vous présenter les résultats obtenus pour la fonte GS 450-10, en les comparant à la recette actuellement employée dans la fonderie.
